# Karnataka Elections 2023 Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sc
```

```python
df=pd.read_csv('bengaluru-voter-turnout-2023.csv')
df.head()
```

```python
df.columns
```

Calculate the total number of voters in each constituency by summing up the "Total Voters" column.

```python
total_voters = df.groupby('Constituency Name')['Total Voters'].sum()
total_voters
```

```python
total_votes_cast = df.groupby('Constituency Name')['Total Voted'].sum()
total_votes_cast
```

Calculate the voter turnout percentage in each constituency by dividing the total number of votes cast by the total number of voters, and multiplying by 100.

```python
voter_turnout = (total_votes_cast / total_voters) * 100
voter_turnout
```

```python

plt.bar(voter_turnout.index, voter_turnout.values)
plt.xticks(rotation=90)
plt.xlabel('Constituency')
plt.ylabel('Voter Turnout (%)')
plt.title('Voter Turnout in Karnataka Elections 2023')
plt.show()
```

```python
df1=pd.read_csv('election-results-kar-may-2023-14_51_51.csv')
df1.head()
```

```python

total_votes_party = df1.groupby(['Constituency', 'Party'])['Votes'].sum()
total_votes_party
```


Calculate the percentage of votes received by each party in each constituency by dividing the total number of votes received by the total number of votes cast in that constituency, and multiplying by 100.

```python
df1[df1['Constituency']=='Tarikere']
```

```python
total_votes_cast_const = df1.groupby('Constituency')['Votes'].sum()
total_votes_cast_const
```

```python
votes_party_percentage = (total_votes_party / total_votes_cast_const) * 100
votes_party_percentage
```

Create a heatmap to visualize the percentage of votes received by each party in each constituency.

```python

votes_party_percentage_pivot = pd.pivot_table(votes_party_percentage.reset_index(), values='Votes', index='Constituency', columns='Party')
sns.heatmap(votes_party_percentage_pivot, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=.5, cbar=False)

# plt.style.set_style('ggplot')
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
plt.xlabel('Party')
plt.ylabel('Constituency')
plt.title('Election Results in Karnataka Elections 2023')
plt.show()
```

```python
election_results_df = pd.read_csv('election-results-kar-may-2023-14_51_51.csv')
election_results_df
```

```python


voter_turnout_df = pd.read_csv('bengaluru-voter-turnout-2023.csv')
election_results_df = pd.read_csv('election-results-kar-may-2023-14_51_51.csv')


merged_df = pd.merge(voter_turnout_df, election_results_df, left_on='Constituency Name', right_on='Constituency')

merged_df['Vote Percentage'] = (merged_df['Votes'] / merged_df['Total Voted']) * 100


sorted_df = merged_df.sort_values(by='Vote Percentage', ascending=False)


for candidate in sorted_df['Candidate'].unique():
    candidate_df = sorted_df[sorted_df['Candidate'] == candidate].head(10)
    print(f"Top 10 Constituencies for {candidate}:")
    for index, row in candidate_df.iterrows():
        print(f"Constituency: {row['Constituency']}")
        print(f"Party: {row['Party']}")
        print(f"Vote Percentage: {row['Vote Percentage']:.2f}%")
        print()
```

```python
# Calculate the overall voter turnout percentage
# Calculate the overall voter turnout percentage
overall_turnout_percentage = (merged_df['Total Voted'].sum() / merged_df['Total Voters'].sum()) * 100
print(f"Overall Voter Turnout Percentage: {overall_turnout_percentage:.2f}%")
```

```python
# Plotting the overall voter turnout percentage by gender
gender_turnout = merged_df.groupby('Total Voters')[['Total Voted Male', 'Total Voted Female', 'Total Voted Other']].sum()
gender_turnout.plot(kind='bar', stacked=True)
plt.xlabel('Total Voters')
plt.ylabel('Total Voted')
plt.title('Voter Turnout by Gender')
plt.legend(loc='upper left')
plt.show()

# Plotting the vote percentage for each candidate
plt.figure(figsize=(10, 6))
sns.barplot(data=sorted_df, x='Candidate', y='Vote Percentage')
plt.xlabel('Candidate')
plt.ylabel('Vote Percentage')
plt.title('Vote Percentage for Each Candidate')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
merged_df['Vote Percentage'] = (merged_df['Votes'] / merged_df['Total Voted']) * 100

# Get the top N candidates with the highest vote percentages
top_candidates = merged_df.groupby('Candidate')['Vote Percentage'].mean().nlargest(10)

# Filter the merged dataframe for the top candidates
top_candidates_df = merged_df[merged_df['Candidate'].isin(top_candidates.index)]

# Plotting the vote percentage for the top candidates
plt.figure(figsize=(10, 6))
sns.barplot(data=top_candidates_df, x='Candidate', y='Vote Percentage')
plt.xlabel('Candidate')
plt.ylabel('Vote Percentage')
plt.title('Vote Percentage for Top Candidates')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
gender_turnout = df[['Total Voted Male', 'Total Voted Female', 'Total Voted Other']].sum()
gender_votes = df[['Total Male Voters', 'Total Female Voters', 'Total Other Voters']].sum()

# Create a DataFrame for the gender-based analysis
gender_analysis = pd.DataFrame({'Gender': ['Male', 'Female', 'Other'],
                                'Turnout': gender_turnout.values,
                                'Votes': gender_votes.values})

# Plot the stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(gender_analysis['Gender'], gender_analysis['Votes'], label='Total Voters', color='lightblue')
plt.bar(gender_analysis['Gender'], gender_analysis['Turnout'], label='Turnout', color='mediumblue')

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender-based Analysis of Voter Turnout and Votes')
plt.legend()

# Show the plot
plt.show()
```
